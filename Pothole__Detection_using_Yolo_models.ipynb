**VISUALIZING THE RESULT OF YOLOv8s MODEL**

```python
import os
import matplotlib.pyplot as plt
import cv2

path = "/content/runs/detect/train5"

images = [
    ("Confusion Matrix", f"{path}/confusion_matrix.png")

]

for title, img_path in images:
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(8,8))
    plt.imshow(img)
    plt.title(title, fontsize=18)
    plt.axis("off")
    plt.show()
```
